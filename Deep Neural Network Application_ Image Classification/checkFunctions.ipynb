```python
import scipy.special as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.zeros((100,1))
```

```python
for i in range(a.shape[0]):
    a[i]=i-50
```

```python
#a = np.linspace(0, 5, 1000)
```

```python
modified_bessel_2nd_kind= sp.jv(0,a)
#print(modified_bessel_2nd_kind)
```

```python
plt.plot(a,modified_bessel_2nd_kind)
print(a.shape)
plt.title("Bessel Function of First Kind")

#plt.ylim(0, 10)
#plt.xlim(-10,10)
plt.show()
```
